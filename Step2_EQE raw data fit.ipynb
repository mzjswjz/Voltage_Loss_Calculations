```python
#loading doesn't flip the data!
df = pd.read_csv(binFile, sep = ',') #uses .csv headers as column labels 
row = df.loc[0]
EQE_folder = row['EQE_folder']
EQE_file = row['EQE_folder'] + '/' + row['EQE_file']
EQEdata = np.genfromtxt(EQE_file, skip_header = 1, delimiter = ',')
EQE_df = pd.DataFrame()
EQE_df['Energy'] = EQEdata[:,2]
EQE_df['EQE'] = EQEdata[:,3]*row['EQE_fac']
os.chdir(codePath) 
phi_bb_df = bb(EQE_df['Energy'])
```

### __Trial option 1. Linear fit__

```python
#'linear' extrapolation

EQE_extrap = extend_EQE(EQE_df, 'linear', min_energy=0.5, start = 1.49, stop = 1.55)
phi_bb_df_etrap = bb(EQE_extrap['Energy'])



plt.figure()
plt.plot(EQE_df['Energy'],EQE_df['EQE'],label = 'Original EQE')
plt.semilogy(EQE_extrap['Energy'],EQE_extrap['EQE'],label = 'linear extrolated EQE')
plt.legend( fontsize=10, frameon=False, loc='best')

# Set x-axis and y-axis labels
plt.xlabel('Energy (eV)')
plt.ylabel('EQE (log)')
```
