### __First test whether current EQE data gives radiative limit of the saturation current density (J$_{0}^{rad}$)__

#### __Change the path for binFile and parameter in block below__

```python
# !!!!!!Change Change path to your bins.csv file before you run this block !!!!!!

binFile = r'/Users/jswjzhm/Desktop/EQE data and analysis/Voltage Loss/240517_Cluster_DCV3T_T1/DV_PHJ_SubNc_DCV3T/bins_Vloss_Linear_EQE.csv'

# !!!!!!!!

# !!!!!!Change this before you run the program !!!!!!

fitting = True # True if you are using extended EQE, False if you are using raw EQE

# !!!!!!!!
```

```python
#loading flips the data!
df = pd.read_csv(binFile, sep = ',') #uses .csv headers as column labels 
row = df.loc[0] #row = df.loc[num] num indicate which line you want to test, 0 = line 1
EQE_file = row['EQE_folder'] + '/' + row['EQE_file']

EQEdata = np.genfromtxt(EQE_file, skip_header = 1, delimiter = ',')
EQE_df = pd.DataFrame()


if fitting == False:
    EQE_df['Energy'] = np.flip(EQEdata[:,2])
    EQE_df['EQE'] = np.flip(EQEdata[:,3])*row['EQE_fac']

else :
    EQE_df['Energy'] = np.flip(EQEdata[:,1])
    EQE_df['EQE'] = np.flip(EQEdata[:,2])*row['EQE_fac']

os.chdir(codePath) 
phi_bb_df = bb(EQE_df['Energy'])
Jsc_EQE = calculate_Jsc(EQE_df['Energy'],EQE_df['EQE'])
J0rad_EQE = J0_rad(EQE_df, phi_bb_df)
vocrad,temp = Voc_rad(row['Voc'], Jsc_EQE, J0rad_EQE)

#cumulative sum (contribution to integral, lower boundary varied for *_rev) 
J0_rad_cum_rev = -q/10*ig.cumulative_trapezoid(np.flip(EQE_df['EQE'].values*phi_bb_df['Phi'].values),np.flip(EQE_df['Energy'].values))
Voc_rad_cum = k*T/q * np.log((Jsc_EQE/J0_rad_cum_rev )+1)
```

### __Plot to see whether J$_{0}^{rad}$ saturated__

```python
## plot Vocrad contributions

display(Math(f'J_{{0}}^{{rad}} \\text{{ is }} {J0rad_EQE}'))
display(Math(f'V_{{oc}}^{{rad}} \\text{{ is }} {vocrad}'))



plt.figure(1)
plt.semilogy(EQE_df['Energy'].values, EQE_df['EQE'].values, label = 'Log(EQE)')
plt.semilogy(EQE_df['Energy'].values, EQE_df['EQE'].values*phi_bb_df['Phi'].values, label = 'Log(EQE* black body emission)')
#plt.ylim([1e-10,1e6])
plt.xlim([0.5,3.5])
plt.ylabel('AU')
plt.xlabel('Energy [eV]')
plt.legend( fontsize=10, frameon=False, loc='best')

plt.figure(2)
plt.semilogy(np.flip(EQE_df.loc[1:,'Energy'].values),J0_rad_cum_rev, label = 'Accumulated J$_{0}^{rad}$')
plt.xlim([0.5,2.5])
plt.ylim([1e-28,1e-14])
plt.ylabel('J$_{0}^{rad}$')
plt.xlabel('Lower boundary of Energy in integral calculating J$_{0}^{rad}$ [eV]')
plt.legend( fontsize=10, frameon=False, loc='best')

plt.figure(3)
plt.plot(np.flip(EQE_df.loc[1:,'Energy'].values),Voc_rad_cum, label = 'Accumulated V$_{0}^{rad}$')
plt.xlim([0.6,3.5])
plt.ylim([0.8,2.5])
plt.ylabel('V$_{0}^{rad}$')
plt.xlabel('Lower boundary of Energy in integral calculating J$_{0}^{rad}$ [eV]')

plt.legend( fontsize=10, frameon=False, loc='best')
```
